```python
# Reading the csv file
df = pd.read_csv('../input/top-1000-imdb-movies/imdb_top_1000.csv')
```

```python
# Checking the overview of the dataframe
df.info()
```

```python
# Selecting a few columns out of all
df = df[['Series_Title','Genre' ,'IMDB_Rating', 'Star1', 'Star2', 'Star3', 'Star4', 'Director', 'Gross']]
```

```python
df.head(3)
```

```python
# Counting the values of genres 
df['Genre'].value_counts()
```

```python
# Checking the most frequent name in the following columns
df['Star1'].value_counts().tail(10)
```

```python
# Checking the movies Tom Hanks has starred in
fil = df['Star1'] == 'Tom Hanks'
df[fil]
df.loc[fil , 'Series_Title']
```

```python
# Check in which columns the name Tom Hanks has been mentioned
df.isin(['Tom Hanks']).any()
# Grabbing all the movies Tom Hanks has acted in
df[df.isin(['Tom Hanks']).any(1)]
# Which genres he has acted in the most
df.loc[df.isin(['Tom Hanks']).any(1), 'Genre'].value_counts()
```

```python
# Another and more lengthy way of grabbing the Tom Hanks columns
df[(df['Star1'] == 'Tom Hanks') | (df['Star2'] == 'Tom Hanks') | (df['Star3'] == 'Tom Hanks') ]
```

```python
# You can also use this for a certain column if you have multiple values
# df[df['Star1'].isin(['Tom Hanks', 'Denzel Washington'])]
# or in the whole df
df[df.isin(['Tom Hanks', 'Denzel Washington']).any(1)]
```

### A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.

```python
df.head()
```

```python
# Check the Data Types
df.dtypes
```

```python
# Group the df by Genre, get the Mean, and sort them based on Gross
df.groupby(['Genre']).mean().sort_values('Gross')
```

```python
genre = df.groupby(['Genre'])
genre.get_group('Family')
```

```python
genre['Star1'].value_counts()
```
